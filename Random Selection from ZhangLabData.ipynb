```python
import os
import pandas as pd
import random
import shutil
```

In this notebook, we are going to randomly select only 490 files for our PNEUMONIA_Bacterial and NORMAL datasets, from unique 490 patients (in other words, one xray per patient). 
We downloaded thousands of Chest XRay files from https://data.mendeley.com/datasets/rscbjbr9sj/3. <br>
<br>
Notes about the data source: <br>
1 - They have separated their files in training and test sets. <br>
2 - Bacterial pneumonia cases also are mixed with Viral pneumonia cases. <br>
3 - Some patients might have more than one xray. <br>
4 - Images are labeled as (disease)-(randomized patient ID)-(image number by this patient).

#### Defining script parameters

```python
# Setting up our random.seed - standard for all scripts
random.seed(42)

# Location of ZhangLabData in local directory
data_directory="C:\\Users\\sofia\\Downloads\\ZhangLabData\\CellData\\chest_xray\\"

# Setting up location of selected files
target_path="C:\\Users\\sofia\\Downloads\\SelectedFiles_from_ZhangLab"
os.mkdir(target_path)
os.mkdir(target_path+"\\NORMAL")
os.mkdir(target_path+"\\PNEUMONIA_Bacterial")

# Specifying the sample size for both Normal and Bacterial Pneumonia
n = 490
```

#### Normal Chest X-Ray files

```python
###########################
#### Listing filenames ####
###########################

########## Train ##########
normal_train_files = os.listdir(data_directory+'train\\NORMAL\\')

# Extract only one xray per patient
normal_train_files_u = [k for k in normal_train_files if '-0001.jpeg' in k] 

normal_train_df = pd.DataFrame({'address': data_directory+'train\\NORMAL\\',
                               'filename': normal_train_files_u})

########### Test ###########
normal_test_files = os.listdir(data_directory+'test\\NORMAL\\')

# Extract only one xray per patient
normal_test_files_u = [k for k in normal_test_files if '-0001.jpeg' in k] 

normal_test_df = pd.DataFrame({'address': data_directory+'test\\NORMAL\\',
                               'filename': normal_test_files_u})

########## All #############
normal_df = normal_train_df.append(normal_test_df)

###########################
##### Selecting files #####
###########################

selected_rows_normal = random.sample(range(0, normal_df.shape[0]), k = n)

## Copying selected files to separate folder
for i in selected_rows_normal:
    shutil.copy(normal_df.iloc[i].address + 
                normal_df.iloc[i].filename,
                target_path+'\\NORMAL')
```

#### Bacterial pneumonia Chest X-Ray files

```python
###########################
#### Listing filenames ####
###########################

########## Train ##########
pneumonia_train_files = os.listdir(data_directory+'train\\PNEUMONIA\\')

# Extract only Bacterial Pneumonia
bacterial_train_files = [k for k in pneumonia_train_files if 'BACTERIA' in k] 

# Extract only one xray per patient
bacterial_train_files_u = [k for k in bacterial_train_files if '-0001.jpeg' in k] 

bacterial_train_df = pd.DataFrame({'address': data_directory+'train\\PNEUMONIA\\',
                               'filename': bacterial_train_files_u})

########## Test ##########

pneumonia_test_files = os.listdir(data_directory+'test\\PNEUMONIA\\')

# Extract only Bacterial Pneumonia
bacterial_test_files = [k for k in pneumonia_test_files if 'BACTERIA' in k] 

# Extract only one xray per patient
bacterial_test_files_u = [k for k in bacterial_test_files if '-0001.jpeg' in k] 

bacterial_test_df = pd.DataFrame({'address': data_directory+'test\\PNEUMONIA\\',
                               'filename': bacterial_test_files_u})

########## All #############
bacterial_df = bacterial_train_df.append(bacterial_test_df)

###########################
##### Selecting files #####
###########################

selected_rows_bacterial = random.sample(range(0, bacterial_df.shape[0]), k = n)

## Copying selected files to separate folder
for i in selected_rows_bacterial:
    shutil.copy(bacterial_df.iloc[i].address + 
                bacterial_df.iloc[i].filename,
                target_path+'\\PNEUMONIA_Bacterial')
```
